```python
#Vamos a mostrar las estadísticas avanzadas:
#PER: El PER (Player Efficiency Rating) es, en esencia, una medida de la productividad por minuto de un jugador.


#¿Cómo se calcula (a grandes rasgos)?
#No es necesaria la fórmula matemática, solo entender qué suma y qué resta:

#Suman: Puntos, rebotes, asistencias, robos, tapones y tiros anotados.

#Restan: Tiros fallados (de campo y libres), pérdidas de balón y faltas personales.

#las 3 claves para entenderlo:
#El "Minuto a Minuto": El PER no premia a alguien solo por jugar mucho tiempo. Si un suplente juega 10 minutos y produce muchísimo, tendrá un PER más alto que un titular que juega 35 minutos pero no hace nada relevante.

#El Promedio es 15: 

#Si tienes un PER de 20, es prácticamente de una estrella.

#Si tienes un PER de 30 o más, estás teniendo una temporada histórica (nivel MVP).

#Ajuste de Equipo: La fórmula tiene un ajuste para que los jugadores de equipos que juegan muy rápido (donde hay más oportunidades de sumar estadísticas) no tengan una ventaja injusta sobre los que juegan en equipos más lentos.

#¿Cuál es su punto débil?
#El PER es muy bueno midiendo la ofensiva, pero se queda corto con la defensa.

#Como la hoja de estadísticas básica solo registra robos y tapones, el PER no puede medir si un jugador defiende bien la posición, si molesta los tiros del rival o si hace una rotación perfecta. Por eso, algunos especialistas defensivos tienen un PER bajo, aunque sean vitales para ganar.
```

```python

df = pd.read_csv('nba_advanced_stats.csv')

# 1. Renombrar columnas para que tengan sentido técnico
columnas_nuevas = {
    'Estación': 'Season',
    'Jugador': 'Player',
    'Equipo': 'Team',
    'Edad': 'Age',
    'Médico de cabecera': 'MP',
    'diputado': 'TS%',
    'PORTS%': 'PER',
    'Gestión de procesos de negocio (BPM)': 'BPM'
}
df.rename(columns=columnas_nuevas, inplace=True)

# 2. Limpiar decimales (cambiar ',' por '.' y convertir a numérico)
cols_numericas = ['TS%', 'USG%', 'PER', 'BPM', 'VORP', 'WS']
for col in cols_numericas:
    if df[col].dtype == 'object':
        df[col] = df[col].str.replace(',', '.').astype(float)

print("Datos listos para el Proyecto_NBA")
print(df.head())
```
